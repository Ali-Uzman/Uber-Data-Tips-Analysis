```python
#importing neccasary module

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#reading or importing data
uber = pd.read_csv("C:\\Users\\uzman\Documents\\python projects & work\\uber-raw-data-sep14.csv")
uber["Date/Time"] = uber["Date/Time"].map(pd.to_datetime)
uber.head()
```

```python
uber["Day"]=uber["Date/Time"].apply(lambda x: x.day)
uber["Weekday"]=uber["Date/Time"].apply(lambda x: x.weekday())
uber["Hour"]=uber["Date/Time"].apply(lambda x: x.hour)
print(uber.head())
```

```python
sns.set(rc= {'figure.figsize':(12,10)})
sns.displot(uber["Day"])
```

```python
sns.displot(uber["Hour"])
```

```python
sns.displot(uber["Weekday"])
```

```python
#Let's observe the correlation of weekday and hour
df=uber.groupby(["Weekday", "Hour"]).apply(lambda x: len(x))
df=df.unstack()
#drawing heatmap using seaborn
sns.heatmap(df, annot= False)
```

```python
#plot density of Uber Trips according to region of Newyork city
uber.plot(kind='scatter', x= "Lon", y= "Lat", alpha= 0.4, s=uber['Day'], 
          label= "Uber Trips", figsize= (12,10), cmap= plt.get_cmap('jet'))
plt.title("Uber Trips Analysis")
plt.legend()
plt.show()
```
